## 2)a)
$$\lim{x → 1} (\frac{x^2-3x+2}{x-1})$$
Fatorando o numerador:
1. Utilizamos o padrão soma-produto;
$$\lim{x → 1} (\frac{(x^2-x)+(-2x+2)}{x-1})$$
2. Em seguida buscamos o maior divisor comum;
$$\lim{x → 1} (\frac{x(x-1)-2(x-1)}{x-1})$$
3. Por fim reescrevemos a função em sua forma fatorada.
$$\lim{x → 1} (\frac{(x-1)(x-2)}{x-1})$$
4. Portanto, o limite da função é igual a -1 quando x se aproxima de 1.
$$\lim_{x → 1} (x-2) = -1$$

## 2)b)
$$\lim{x → 2} (\frac{3x^2-2x-5}{x-1})$$
Fatorando o numerador:
1. O primeiro passo consiste em dividir o termo bx em dois termos;
$$\lim{x → 2} (\frac{3x^2-5x+3x-5}{x-1})$$
2. Em seguida precisamos fatorar os dois primeiros e os dois últimos termos de forma independente;
$$\lim{x → 2} (\frac{x(\frac{3x^2}{x}-\frac{5x}{x})+(3x-5)}{x-1})$$
3. Nós precisamos reduzir as frações para seus menores termos possíveis;
$$\lim{x → 2} (\frac{x(3x-5)+(3x-5)}{x-1})$$
4. É necessário descobrir o máximo divisor comum do função;
$$\lim{x → 2} (\frac{(3x-5) (\frac{x(3x-5)}{3x-5}-\frac{3x-5}{3x-5})}{x-1})$$
5. Reduzindo a fração para seus menores termos;
$$\lim{x → 2} (\frac{(3x-5)(x+1)}{x-1})$$
6. Quando x tende a 2 temos;
$$\lim{x → 2} (\frac{(3*2-5)(2+1)}{2-1})$$
$$\lim{x → 2} (\frac{(6-5)(3)}{1})$$
$$\lim_{x → 2} 1*3 = 3$$
7. Portanto, o limite da função é igual a 3 quando x se aproxima de 2.

## 3)a)
$$\lim{x → 9} \sqrt{x-8}+3$$
Quando x tende a 9 temos:
$$\lim{x → 9} \sqrt{9-8}+3$$
$$\lim{x → 9} \sqrt{1}+3$$
$$\lim{x → 9} 1+3=4$$
Portanto, o limite da função é igual a 4 quando x se aproxima de 9.

```python
x = np.linspace(-10, 15, 256, endpoint = True)
y = np.sqrt(x-8)+3

fig = plt.figure(figsize = (10,8))
plt.plot(x, y, '-b')

x = sp.Symbol('x')
expr = sp.sqrt(x-8) + 3

limite = sp.limit(expr, x, 9)
print("Limite da expressão quando tende a 9 : {}".format(limite))


plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Logaritmica')
plt.savefig('logaritmicaDecrescente.png', dpi=300)
plt.xlim([-5,15])
plt.ylim([-5,15])
plt.show()
```

## 3)b)
$$\lim_{x → \frac{3}{2}} (\frac{2x^2-5x+3}{2x^2-3x})$$
Fatorando a função:
1. O primeiro passo consiste em dividir o termo x em dois termos;
$$\lim_{x → \frac{3}{2}} (\frac{2x^2-3x-2x+3}{2x^2-3x})$$
2. Em seguida precisamos fatorar os dois primeiros e os dois últimos termos de forma independente;
$$\lim_{x → \frac{3}{2}} (\frac{x(\frac{2x^2}{x}-\frac{3x}{x})+(-2x+3)}{x(\frac{2x^2}{x}-\frac{3x}{x})})$$
3. Nós precisamos reduzir as frações para seus menores termos possíveis;
$$\lim_{x → \frac{3}{2}} (\frac{x(2x-3)+(-2x+3)}{x(2x-3)})$$
4. É necessário descobrir o máximo divisor comum do função;
$$\lim_{x → \frac{3}{2}} (\frac{(2x-3)(\frac{x(2x-3)}{2x-3}-\frac{2x+5}{2x-3})}{x(2x-3)})$$
5. Reduzindo a fração para seus menores termos;
$$\lim_{x → \frac{3}{2}} (\frac{(2x-3)(x-1)}{x(2x-3)})$$
6. Cancelamos os fatores comuns;
$$\lim_{x → \frac{3}{2}} (\frac{x-1}{x})$$
7. Quando x tende a 3/2 temos;
$$\lim_{x → \frac{3}{2}} (\frac{\frac{3}{2}-1}{\frac{3}{2}})$$
$$\lim_{x → \frac{3}{2}} (\frac{\frac{1}{2}}{\frac{3}{2}})=\frac{1}{3}$$
8. Portanto, o limite da função quando x tende a 3/2 é igual a 1/3.

```python
x = np.linspace(-10, 10, 256, endpoint = False)
y = (2*x**2-5*x+3)/(2*x**2-3*x)

fig = plt.figure(figsize = (10,8))
plt.plot(x, y, '-b')

x = sp.Symbol('x')
expr = (2*x**2-5*x+3)/(2*x**2-3*x)

limite = sp.limit(expr, x, 1.5)
print("Limite da expressão quando tende a 3/2 : {}".format(limite))



plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Logaritmica')
plt.savefig('logaritmicaDecrescente.png', dpi=300)
plt.xlim([-10,10])
plt.ylim([-10,10])
plt.show()
```

## 3)c)
$$\lim_{x → 9} (\frac{\sqrt{x}-3}{x-9})$$
Simplificando função:
1. Para simplificar a expressão, podemos multiplicar o numerador e o numerador pelos seus conjugados;
$$\lim_{x → 9} (\frac{\sqrt{x}-3}{x-9})(\frac{\sqrt{x}+3}{\sqrt{x}+3})$$
$$\lim_{x → 9} (\frac{\sqrt{x}^2-3^2}{(x-9)(\sqrt(x)+3)})$$
$$\lim_{x → 9} (\frac{x-9}{(x-9)*(\sqrt{x}+3)})$$
$$\lim_{x → 9} (\frac{1}{(\sqrt{x}+3)})$$
2. Quando x tende a 9 temos;
$$\lim_{x → 9} (\frac{1}{(\sqrt{9}+3)})$$
$$\lim_{x → 9} (\frac{1}{6})$$
3. Portanto, o limite da função quando x tende a 9 é igual a 1/6.

```python
x = np.linspace(-0.5, 10, 256, endpoint = True)
y = (np.sqrt(x)-3)/(x-9)

fig = plt.figure(figsize = (10,8))
plt.plot(x, y, '-b')

x = sp.Symbol('x')
expr = (sp.sqrt(x)-3)/(x-9)

limite = sp.limit(expr, x, 9)
print("Limite da expressão quando tende a 9 : {}".format(limite))


plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Função Logaritmica')
plt.savefig('logaritmicaDecrescente.png', dpi=300)
plt.xlim([0,0.5])
plt.ylim([0,0.5])
plt.show()
```

## **4)a)**
a)
$$\lim_{x \to \infty} \frac{x-3}{7-x}$$
1) Dividimos pelo denominador de maior potência:
$$\frac{1 - \frac{3}{x}}{\frac{7}{x} - 1}$$
2) Definimos os limites de cada parte da fração:
$$\lim_{x \to \infty} 1 - \lim_{x \to \infty} \frac{3}{x} = 1 - 0 = 1$$
$$\lim_{x \to \infty} \frac{7}{x} - \lim_{x \to \infty} -1 = 0 - 1 = -1$$
3) simplificamos:
$$\lim_{x \to \infty} \frac{1}{-1} = -1$$

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import *

x = symbols('x')
expr = (x-3)/(7-x);
print("Expressão: {}".format(expr))
limit_expr = limit(expr, x, float('inf'))
print("Limite da expressão: {}".format(limit_expr))

x = np.linspace(0, 10, 256, endpoint = False)
y = (x-3)/(7-x)

fig = plt.figure(figsize = (10,8))
plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## **4)b)**
b) $$\lim_{x \to \infty} \frac{x-1}{x²+5x-6}$$
1) Fatoramos a seguinte equação:
$$x²+5x-6=(x-1)(x+6)$$
2) Simplificamos o resultado:
$$\lim_{x \to \infty} \frac{x-1}{(x-1)(x+6)} = \frac{1}{(x+6)}$$
3) Descobrimos os limites:
$$\lim_{x \to \infty} 1 = 1$$
$$\lim_{x \to \infty} (x+6) = \infty$$
$$\lim_{x \to \infty} \frac{1}{\infty} = 0$$

```python
x = symbols('x')
expr = (x-1)/(x**2 + 5*x - 6);
print("Expressão: {}".format(expr))
limit_expr = limit(expr, x, float('inf'))
print("Limite da expressão: {}".format(limit_expr))

x = np.linspace(-4, 10, 256, endpoint = False)
y = (x-1)/(x**2 + 5*x - 6);

fig = plt.figure(figsize = (10,8))
plt.plot(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
